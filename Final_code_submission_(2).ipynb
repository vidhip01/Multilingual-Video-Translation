EXTRACTING AUDIO FROM VEDIO

````
#extracting audio from video youtube link
video='https://www.youtube.com/shorts/HYJZ0CYw3hk'
data=pytube.YouTube(video)
audio=data.streams.get_audio_only()
A=audio.download()
Aud=A[9:]
Aud
````

TRANSCRIBING AUDIO USING WHISPER

````
# Installing Whisper libary
!pip install git+https://github.com/openai/whisper.git -q
import whisper
````

````
#Loading large model
model1 = whisper.load_model('large')
decode_options={"language":"english"}
text = model1.transcribe(Aud,word_timestamps= True,verbose=True) #extracting time stamps for later audio synching
#here, model is detecting the language on it's own and transcribing with time stamps
````

TRANSLATING

````
#translating from hinglish to english on the basis of pauses taken by speaker in the vedio
decode_options={"language":"english"}
text_eng = model1.transcribe(Aud,**decode_options,word_timestamps= True,verbose=True)
````

````
#translation to telugu using googletranslate
!pip install googletrans==3.1.0a0
import googletrans
from googletrans import Translator
translator=Translator()
````

````
text_eng['segments'][0] #checking what information, we have inside text_eng, we have useful info in segments
````

````
text_eng_segments=text_eng['segments'] #extracting segment portion of text_eng
text_eng_segments[0]['text']
````

````
#translating each eng phrase to telugu in this list of dictionaries. Text is inside 'text' key of each dictionary
i=0
for i in range(len(text_eng_segments)):
  for key in text_eng_segments[i]:
    if key=='text':
      text_eng_segments[i]['text']=translator.translate(text_eng_segments[i]['text'],src='en',dest='te').text #translating to telugu
````

STORING TRANSLATIONS, TIMESTAMS IN LIST

````
new_list_tel=[] #new list creation for telugu translations
new_list_time_start=[] #for storing starting time of these phrases
new_list_time_end=[] #for storing ending time of these phrases
````

````
for i in range(len(text_eng_segments)):
  new_list_tel.append(text_eng_segments[i]['text'])
  new_list_time_start.append(text_eng_segments[i]['start'])
  new_list_time_end.append(text_eng_segments[i]['end'])
````

TEXT TO SPEECH

````
!pip install gTTS  #installing gTTS for text to speech
#Importing libraries

from gtts import gTTS
import os
````

````
#text to speech
audio_list_tel=[]
for i in range(len(new_list_tel)) :
  tel_audio='tel_audio.mp3 '+str(i)
  language='te'
  sp=gTTS(text=new_list_tel[i], lang=language, slow=False)
  sp.save(tel_audio)
  audio_list_tel.append(tel_audio)
````

````
audio_list_tel[:5] #list of telugu translated audios created
````

AUDIO ALIGNMENT

````
#for audio manipulation
!pip install pydub
import pydub
from pydub import AudioSegment
````

````
from decimal import Decimal #for accurate arithematic
````

````
sound = AudioSegment.from_mp3(audio_list_tel[1])
sound.duration_seconds
````

````
for i in range(len(audio_list_tel)) :
  sound = AudioSegment.from_mp3(audio_list_tel[i])
   #aligning start and end time of each phrase
  play_back_speed= float(sound.duration_seconds)/float(float(new_list_time_end[i])-float(new_list_time_start[i]))
  sound=sound.speedup(playback_speed=float(play_back_speed))
  sound.export(audio_list_tel[i], format="mp3")
````

````
#adding pauses
combined_sounds = AudioSegment.empty()
for i in range(len(audio_list_tel)) :
  if i==0:
     one_sec_segment = AudioSegment.silent(duration=float(new_list_time_start[i])*float(1000))
  else:
     one_sec_segment = AudioSegment.silent(duration=float(float(new_list_time_start[i])-float(new_list_time_end[i-1]))*float(1000))
  sound = AudioSegment.from_file(audio_list_tel[i])
  combined_sounds=combined_sounds+sound+one_sec_segment
  combined_sounds.export("combined.mp3", format="mp3")
````
